```python
from scipy.stats import norm

mean = 64.43
std_dev = 2.99

x = norm.cdf(61.43, mean, std_dev)

print(x)
```

```python
x = norm.ppf(0.975, mean, std_dev)

print(x)
```

```python
import random

for i in range(0,1000):
    random_p = random.uniform(0.0, 1.0)
    random_weight = norm.ppf(random_p, loc=64.43, scale=2.99)
    print(random_weight)
```

```python
mean = 18

std_dev = 1.5
x = norm.cdf(19.5, mean, std_dev) - norm.cdf(16.5, mean, std_dev)
print(x)
x = norm.cdf(21, mean, std_dev) - norm.cdf(15, mean, std_dev)
print(x)
x = norm.cdf(22.5, mean, std_dev) - norm.cdf(13.5, mean, std_dev)
print(x)
```
