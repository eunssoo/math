```python
from sympy import *
import numpy as np

# 로지스틱 그래프 그리기
x = symbols('x')

a = 1
b = 3

logistic = 1 / (1 + exp(-(a*x + b)))

plot(logistic)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -10부터 10까지 0.1 간격
x = np.arange(-10, 10.1, 0.1)

logistic = 1 / (1 + np.exp(-x))

plt.plot(x, logistic)
plt.grid(True)
plt.show()
```
