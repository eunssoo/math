```python
sample = [0, 1, 5, 7, 9, 10, 14]
```

```python
def median(values):
    ordered = sorted(values)
    print(ordered)
    n = len(ordered)
    mid = int(n/2) - 1 if n %2 == 0 else int(n/2)

    if n %2 == 0:
        return (ordered[mid] + ordered[mid+1]) /2.0
    else:
        return ordered[mid]
print(median(sample))
```

```python
def median(values):                       # median이라는 함수를 만든다. values는 여러 숫자가 들어있는 리스트라고 보면 된다.
    ordered = sorted(values)              # values를 작은 수부터 큰 수 순서로 정렬해서 ordered에 저장한다.
    print(ordered)                        # 정렬된 결과를 화면에 출력한다.
    n = len(ordered)                      # 정렬된 리스트의 전체 길이(원소 개수)를 n에 저장한다.
    mid = int(n/2) - 1 if n % 2 == 0 else int(n/2)   # 원소 개수가 짝수면 가운데 왼쪽 인덱스, 홀수면 정확한 가운데 인덱스를 구한다.

    if n % 2 == 0:                        # 원소 개수가 짝수라면
        return (ordered[mid] + ordered[mid+1]) / 2.0 # 가운데 두 값의 평균을 반환한다.
    else:                                 # 원소 개수가 홀수라면
        return ordered[mid]               # 가운데 값 하나를 그대로 반환한다.

print(median(sample))                     # sample 리스트의 중앙값을 구해서 출력한다.
```
